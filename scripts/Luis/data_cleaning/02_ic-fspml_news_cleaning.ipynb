```python
import pandas as pd
```

```python
data_path = [
    "data/raw/ic-fspml_stock_news_sentiment_train.csv",
    "data/raw/ic-fspml_stock_news_sentiment_validation.csv",
    "data/raw/ic-fspml_stock_news_sentiment_test.csv",
]
```

```python
df = pd.DataFrame()

for file in data_path:
    df = pd.concat(
        [
            df,
            pd.read_csv(file)
        ]
    )
```

```python
# Name is suppossed to be company name for the ticker it repesents but there is Nan however ticker has no nans, lets look into it
display(df.isna().sum())

display(df.loc[df["name"].isna()])
```

```python
# None of the tickers with no name have a name in one row but none in any other so we will leave the unnamed as is.
for ticker in df.loc[df["name"].isna()]["ticker"].unique():
    print(df.loc[df["ticker"] == ticker]["name"].value_counts())
```

```python
# Finally we will filter the tickers that we look from in our data
tickers_df = pd.read_csv("data/processed/NASDAQ_stock_data_processed.csv", index_col=0)

ticker_list = []
display(tickers_df.head(2))
for i in range(0,len(tickers_df),6):
    ticker_list.append(tickers_df.columns[i].split("_")[0])

print(ticker_list)
```

```python
in_df_not_in_tickers = list(set(df.ticker) - set(ticker_list))
print(f"Tickers in df that are not in our required tickets ({len(in_df_not_in_tickers)}):")
print(in_df_not_in_tickers)

print("###########################################################################################################################################")
print("###########################################################################################################################################")
print("###########################################################################################################################################")
in_tickers_not_in_df = list(set(ticker_list) - set(df.ticker))
print(f"Tickers in required tickers not in df ({len(in_tickers_not_in_df)} out of {len(tickers_df)}):")
print(in_tickers_not_in_df)
```

```python
df["article_date"] = pd.to_datetime(df["article_date"])
df = df.sort_values(by="article_date")
# We filter dates outside of our investigation scope
df = df.loc[df["article_date"].between("2012-01-03", "2020-12-31")]
df
```

```python
# Filter the news for our tickers
df = df.loc[df["ticker"].isin(ticker_list)]
df.reset_index(drop=True, inplace=True)
df.to_csv("data/processed/ic-fspml_stock_news_sentiment_processed.csv", index=False)
```
