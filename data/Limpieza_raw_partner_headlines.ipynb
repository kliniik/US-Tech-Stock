```python
import pandas as pd
```

```python
partner_headlines = pd.read_csv("raw/raw_partner_headlines.csv")
partner_headlines.head()
```

```python
print(sum(partner_headlines["headline"].isna()))
print(sum(partner_headlines["date"].isna()))
print(sum(partner_headlines["stock"].isna()))
```

No hay ninguna columna que nos interese con NAs, por lo que simplemente nos quedamos con las columnas *headline*, *date* y *stock*, corregimos la columna *Unnamed: 0*, le cambiamos el nombre a *index* y se la asignamos al índice del dataframe.

```python
partner_headlines["index"] = list(range(len(partner_headlines)))
partner_headlines = partner_headlines[["index", "headline", "date", "stock"]]
partner_headlines = partner_headlines.set_index("index")
```

```python
partner_headlines
```

```python
partner_headlines.iloc[:len(partner_headlines)//2,].to_csv("processed/partner_headlines_cleaned-1.csv")
partner_headlines.iloc[len(partner_headlines)//2:,].to_csv("processed/partner_headlines_cleaned-2.csv")
```
