```python
import pandas as pd
```

```python
file = "data/raw/David-Chew-HL_Tech-Stocks-News_train.csv"
df = pd.read_csv(file)
display(df)
```

```python
df.isna().sum()
```

```python
df["date"] = pd.to_datetime(df["date"])
df = df.sort_values(by="date")
df.reset_index(drop=True,inplace=True)
# Since news go up to 2024 we can filter out the dates we don't need
df
```

```python
print("Is every headline unique?",df.shape[0] == df["headline"].nunique())
```

```python
# Search duplicate headlines
df.loc[df['headline'].duplicated(keep=False)].sort_values('headline')
```

```python
# Is there any duplicate headlines with different date?
df_duplicated_diff_date = df.loc[df.duplicated(subset=['headline'], keep=False) & ~df.duplicated(subset=['date', 'headline'], keep=False)].sort_values("headline")

display(df_duplicated_diff_date)
display(df_duplicated_diff_date.shape)
```

```python
# Since duplicate in different date can be similiar to a "repost" of a news article and posting it in a different date may cause it to have a different impact, we will keep duplicate articles which have a duplicate in different date but not the duplicates which are all duplicate in the same date

idx_duplicates = df.loc[df["headline"].duplicated()].index
idx_duplicates

idx_duplicates_diff_date = df_duplicated_diff_date.index

drop_index = idx_duplicates.difference(idx_duplicates_diff_date)

print("df shape", df.shape)
print("Duplicates shape",idx_duplicates.shape)
print("Duplicates in diff date shape",idx_duplicates_diff_date.shape)
print("Duplicates in same date shape",drop_index.shape)

print("Before df shape", df.shape)
print("After df shape", df.drop(drop_index).shape)
```

```python
df = df.drop(drop_index).reset_index(drop=True)
```

```python
df = df.sort_values(by=["date","headline"]).reset_index(drop=True)
df
```
