```python
# ordenamos y extraemos las listas
y = []
preds = [preds[i] for i in (0, 7, 9, 2, 8, 10, 1, 3, 4, 5, 6)]
for i in preds:
    for j in i:
        y.extend(j)
print(len(preds))
print(len(y))
```

```python
dataset_path = "./full_dataset_removed_duplicates.csv"  # este es el dataset completo pero quitando las noticias que aparecían repetidas en una misma celda

df = pd.read_csv(dataset_path, encoding="utf-8", encoding_errors="replace").iloc[:, 1:]
df
```

```python
mask = df.apply(lambda col: col.str.contains(r"noticias"))

indices = np.argwhere(mask.values == True).tolist()
indices = list(map(tuple, indices))  # indices de las celdas que tienen noticias en df

noticias_con_sentimiento = []

for i in indices:
    cell = ast.literal_eval(df.iloc[i])
    for x in range(len(cell["noticias"])):
        if cell["noticias"][x]["sentimiento"] is not None:
            noticias_con_sentimiento.append(((i, x), cell["noticias"][x]["sentimiento"]))
```

```python
# hay noticias que ya tenían sentimiento, pero el sentimiento es distinto a positive, neutral o negative, así que descartamos
noticias_con_sentimiento_copy = noticias_con_sentimiento.copy()
sentimientos = ("neutral", "positive", "negative")
for noticia in noticias_con_sentimiento_copy:
    if noticia[1] not in sentimientos:
        noticias_con_sentimiento.remove(noticia)
print(noticias_con_sentimiento[:5])
print(len(noticias_con_sentimiento_copy), len(noticias_con_sentimiento))
```

```python
print(set(map(lambda x: x[1], noticias_con_sentimiento)))
```

```python
# este dataframe contiene las noticias que ya tenían sentimiento
# lo guardamos para comparar después con los sentimientos que hemos sacado nosotros
df_noticias_con_sentimiento = pd.DataFrame({"idx": list(map(lambda x: x[0], noticias_con_sentimiento)), "sentimiento": list(map(lambda x: x[1], noticias_con_sentimiento))})
```

```python
# añadimos el sentimiento al dataset completo
# otra vez, debido a la gran cantidad de datos se realizó en tres máquinas simultáneamente
try:
    for i in range(len(preds_dataset)-1, 1389000, -1):
        if i%1000==0:
            print(i)
        df_indexes = eval(preds_dataset["idx"][i])
        for idx in df_indexes:
            cell = eval(df.iloc[idx[0]])
            cell["noticias"][idx[1]]["sentimiento"] = preds_dataset["pred"][i]
            df.iloc[idx[0]] = str(cell)
except KeyboardInterrupt:
    print(i)
    pickle.dump(df, open(f"dataset_con_sentimiento_{i}-{len(preds_dataset)}.pkl", "wb"))
```
