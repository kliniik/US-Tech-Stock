```python
df = pd.read_csv("data/processed/NASDAQ_stock_data_processed.csv")

def make_cell_into_dict(row):
    return [row[0]] + [{"val_adj_close":val} for val in row[1:]]

df = pd.DataFrame(list(map(make_cell_into_dict, df.values)), columns=df.columns)
df.head()
```

```python
DATE_COL = "date"
SENTIMENT_COL = "sentiment"
HEADLINE_COL = "headline"
TICKER_COL = "ticker"
```

```python
TICKERS = df.columns.tolist()[1:]
```

```python
def find_date_column(columns):
    possible_date_col = ["date", "created_at","post_date","article_date"]
    for col in possible_date_col:
        if col in columns:
            return col
    
    raise ValueError("No date column found")

def find_ticker_column(columns):
    possible_ticker_col = ["ticker", "stock","ticker_symbol"]
    for col in possible_ticker_col:
        if col in columns:
            return col
    
    raise ValueError("No ticker column found")

def find_headline_column(columns):
    possible_headline_col = ["text","headline","body","article_headline","title"]
    for col in possible_headline_col:
        if col in columns:
            return col
    
    raise ValueError("No headline column found")

def find_label_column(columns):
    possible_label_col = ["label"]
    for col in possible_label_col:
        if col in columns:
            return col

    return None

def find_cols(columns):
        date_col = find_date_column(columns)
        ticker_col = find_ticker_column(columns)
        headline_col = find_headline_column(columns)
        label_col = find_label_column(columns)
        
        return date_col, ticker_col, headline_col, label_col

def join_parts_df(file_paths: list[str]) -> pd.DataFrame:
    def print_timestamp(msg):
        print(f"[{datetime.now().strftime('%Y-%m-%d %H:%M:%S.%f')}] {msg}")

    print_timestamp("Starting join_parts_df")
    print("file_paths:", file_paths)

    print_timestamp("Starting DataFrame concatenation")
    df = pd.DataFrame()

    for file_path in file_paths:
        print_timestamp(f"Reading file: {file_path}")
        if ("tweets_remaining_cleared" in file_path) or ("tweets_labelled_cleared" in file_path):
            sep = ";"
        else:
            sep = ","
        df = pd.concat([df, pd.read_csv(file_path, sep=sep)])

    print_timestamp("Finished DataFrame concatenation")
    
    print_timestamp("Starting column cleanup")
    if "index" in df.columns:
        df = df.drop(columns=["index"])
    if "id" in df.columns:
        df = df.drop(columns=["id"])

    date_col, ticker_col, headline_col, label_col = find_cols(df.columns)
    print("df columns:", df.columns)
    print("date_col:", date_col)
    print("ticker_col:", ticker_col)
    print("headline_col:", headline_col)
    print("label_col:", label_col)

    print_timestamp("Starting DataFrame transformation")
    print("Transforming df...")

    if label_col:
        df = df.rename(columns={headline_col: HEADLINE_COL,
                            date_col: DATE_COL,
                            ticker_col: TICKER_COL,
                            label_col: SENTIMENT_COL
                            })
    else:
        df = df.rename(columns={headline_col: HEADLINE_COL,
                            date_col: DATE_COL,
                            ticker_col: TICKER_COL
                            })

    print_timestamp("Converting dates")
    df[DATE_COL] = pd.to_datetime(df[DATE_COL], utc=True).dt.normalize()
    
    print_timestamp("Sorting values")
    df = df.sort_values(by=[DATE_COL])
    
    print_timestamp("Filtering date range")
    df = df.loc[df["date"].between(pd.to_datetime(START_DATE, utc=True), 
                                  pd.to_datetime(END_DATE, utc=True))]
    print("Filtering columns")
    if label_col:
        df = df.loc[:, [DATE_COL, TICKER_COL, HEADLINE_COL, SENTIMENT_COL]]
    else:
        df = df.loc[:, [DATE_COL, TICKER_COL, HEADLINE_COL]]

    print("Took", df.columns.values, "cols")

    print("Filtering by ticker")
    df = df.loc[df[TICKER_COL].isin(TICKERS)]

    df = df.reset_index(drop=True)

    print_timestamp("Creating pivot table")
    if label_col:

        df["headline_setnames"] = list(zip(df[HEADLINE_COL], df[SENTIMENT_COL]))
        df = df.drop(columns=[HEADLINE_COL, SENTIMENT_COL])

        df = df.pivot_table(
            index=DATE_COL,
            columns=TICKER_COL,
            values="headline_setnames",
            aggfunc=lambda x: list(x),
        )
        has_label = True
    else:
        df = df.pivot_table(
            index=DATE_COL,
            columns=TICKER_COL,
            values=HEADLINE_COL,
            aggfunc=lambda x: list(x),
        )
        has_label = False

    print_timestamp("Finished join_parts_df")
    return df,has_label
```

```python
from tqdm.notebook import tqdm
import pandas as pd

def update_dataframe_with_news(df, groups_of_datasets):
    # Create date and ticker mappings
    date_to_idx = {pd.to_datetime(date).strftime('%Y-%m-%d'): idx 
                   for idx, date in enumerate(df["Date"].values)}
    ticker_to_idx = {ticker: idx for idx, ticker in enumerate(df.columns.values)}
    
    outer_p_bar = tqdm(total=len(groups_of_datasets), desc="Processing groups")
    
    for group in groups_of_datasets:
        outer_p_bar.update(1)
        dataset, has_label = join_parts_df(group)  # Assuming this function exists
        
        print("Starting to process news into DataFrame")
        
        progress_total = dataset.size
        p_bar = tqdm(total=progress_total, desc="Processing entries")
        
        for i in range(dataset.shape[0]):
            for j in range(dataset.shape[1]):
                p_bar.update(1)
                
                # Skip if the cell is empty or NaN
                try:
                    if pd.isna(dataset.iloc[i,j]):
                        continue
                except ValueError:
                    
                    try:
                        ticker = dataset.columns[j]
                        fecha = dataset.index[i]
                        
                        # Find the closest matching date within 7 days
                        fecha_str = pd.to_datetime(fecha).strftime('%Y-%m-%d')
                        days_searched = 0
                        while fecha_str not in date_to_idx and days_searched < 7:
                            fecha = fecha + pd.Timedelta(days=1)
                            fecha_str = pd.to_datetime(fecha).strftime('%Y-%m-%d')
                            days_searched += 1
                            
                        if fecha_str not in date_to_idx:
                            print(f"WARNING: Could not find matching date for {dataset.index[i]} within 7 days")
                            continue
                            
                        idx_date_en_df = date_to_idx[fecha_str]
                        idx_ticker_en_df = ticker_to_idx[ticker]
                        
                        # Get current dictionary and ensure it's properly formatted
                        try:
                            current_value = df.iloc[idx_date_en_df, idx_ticker_en_df]
                            current_dict = eval(current_value) if isinstance(current_value, str) else {}
                        except (ValueError, SyntaxError, TypeError):
                            current_dict = {}
                        
                        # Initialize news list if it doesn't exist
                        if "noticias" not in current_dict:
                            current_dict["noticias"] = []
                        
                        # Add news items based on whether they have sentiment labels
                        news_items = dataset.iloc[i,j]
                        if has_label:
                            for noticia, sentimiento in news_items:
                                current_dict["noticias"].append({
                                    "noticia": noticia,
                                    "sentimiento": sentimiento
                                })
                        else:
                            for noticia in news_items:
                                current_dict["noticias"].append({
                                    "noticia": noticia,
                                    "sentimiento": None
                                })
                        
                        # Update the DataFrame with the modified dictionary
                        df.iloc[idx_date_en_df, idx_ticker_en_df] = str(current_dict)
                        
                    except Exception as e:
                        print(f"Error processing entry at {i},{j}: {str(e)}")
                        continue
                    
        p_bar.close()
        
    outer_p_bar.close()
    return df

# Usage:
df = update_dataframe_with_news(df, groups_of_datasets)
```
